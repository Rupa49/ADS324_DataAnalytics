```python
import numpy as np
import matplotlib.pyplot as plt

# Load image as NumPy array
image = plt.imread("/content/Cartoon.jpeg")

print(image.shape)   # (H, W, C)
```

```python
h_flip = np.fliplr(image)

plt.imshow(h_flip)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

```python
v_flip = np.flipud(image)

plt.imshow(v_flip)
plt.title("Vertical Flip")
plt.axis("off")
plt.show()
```

```python
rotate_90 = np.rot90(image)

plt.imshow(rotate_90)
plt.title("Rotate 90°")
plt.axis("off")
plt.show()
```

```python
rotate_180 = np.rot90(image, 2)

plt.imshow(rotate_180)
plt.title("Rotate 180°")
plt.axis("off")
plt.show()
```

```python
bright = np.clip(image + 50, 0, 255).astype(np.uint8)

plt.imshow(bright)
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```

```python
dark = np.clip(image - 50, 0, 255).astype(np.uint8)

plt.imshow(dark)
plt.title("Brightness Decreased")
plt.axis("off")
plt.show()
```

```python
noise = np.random.normal(0, 25, image.shape)
noisy = np.clip(image + noise, 0, 255).astype(np.uint8)

plt.imshow(noisy)
plt.title("Gaussian Noise")
plt.axis("off")
plt.show()
```

```python
h, w, _ = image.shape
cropped = image[h//4:3*h//4, w//4:3*w//4]

plt.imshow(cropped)
plt.title(" Resize Cropped Image")
plt.axis("off")
plt.show()
```

```python
scaled = image[::2, ::2]

plt.imshow(scaled)
plt.title("Scaled Down")
plt.axis("off")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = plt.imread("/content/Cartoon.jpeg")

# Convert to grayscale (if RGB)
if image.ndim == 3:
    gray = np.mean(image, axis=2).astype(np.uint8)
else:
    gray = image

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()


```

```python
threshold = 128

binary = np.where(gray > threshold, 255, 0).astype(np.uint8)

plt.imshow(binary, cmap="gray")
plt.title("Binary Thresholding")
plt.axis("off")
plt.show()
```
