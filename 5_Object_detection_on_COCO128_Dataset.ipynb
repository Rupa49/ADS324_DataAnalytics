```python
import random
import os
import cv2
import matplotlib.pyplot as plt

# The 'folder' variable should be the path to the directory containing images, not an image file itself.
# Corrected path to the directory
image_directory = "/content/coco128/images/train2017"

# Get a list of all files in the directory
image_files = [f for f in os.listdir(image_directory) if os.path.isfile(os.path.join(image_directory, f))]

# Choose a random image file from the list
img_filename = random.choice(image_files)
full_path = os.path.join(image_directory, img_filename)

print("Image:", full_path)

results = model(full_path)

img = results[0].plot()
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis("off")
plt.show()
```
