```python
import spacy

# Load the English model
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    print("SpaCy model 'en_core_web_sm' not found. Please run '!python -m spacy download en_core_web_sm' first.")
    exit()

text_spacy = "Apple is looking at buying U.K. startup for $1 billion."
doc = nlp(text_spacy)

print("\n--- SpaCy Tokenization, POS Tagging, and NER ---")

# Tokenization, POS Tagging, and Lemmatization
print("\nTokens, POS, Lemma, and Dependencies:")
for token in doc:
    print(f"{token.text:{10}} {token.pos_:{8}} {token.lemma_:{10}} {token.dep_:{10}}")

# Named Entity Recognition
print("\nNamed Entities:")
for ent in doc.ents:
    print(f"{ent.text:{15}} {ent.label_:{10}} {spacy.explain(ent.label_)}")
```

### 1. Tokenization
Tokenization is the process of breaking down a piece of text into smaller units called 'tokens'. These tokens can be words, punctuation marks, or even sentences.

```python
from nltk.tokenize import word_tokenize, sent_tokenize

text = "Natural Language Processing is a fascinating field. It combines computer science and artificial intelligence."

# Word Tokenization
words = word_tokenize(text)
print("Word Tokens:", words)

# Sentence Tokenization
sentences = sent_tokenize(text)
print("Sentence Tokens:", sentences)
```

### 2. Stemming and Lemmatization
**Stemming** reduces words to their root or stem, which may not be a dictionary word (e.g., 'running' -> 'run'). **Lemmatization** reduces words to their base or dictionary form (lemma), which is always a valid word (e.g., 'better' -> 'good').

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words]
print("Stemmed Words:", stemmed_words)

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print("Lemmatized Words:", lemmatized_words)

# Example with POS tag for better lemmatization
print("Lemmatizing 'running' as verb:", lemmatizer.lemmatize("running", pos="v"))
print("Lemmatizing 'better' as adjective:", lemmatizer.lemmatize("better", pos="a"))
```

### 3. Part-of-Speech (POS) Tagging
POS tagging involves assigning a grammatical category (e.g., noun, verb, adjective) to each word in a text.

```python
from nltk import pos_tag

pos_tags = pos_tag(words)
print("POS Tags:", pos_tags)
```

### 4. Named Entity Recognition (NER)
NER identifies and classifies named entities in text into predefined categories such as person names, organizations, locations, expressions of times, quantities, monetary values, percentages, etc.

```python
from nltk import ne_chunk

# We need POS tags for NER
pos_tokens = pos_tag(word_tokenize("Google's CEO Sundar Pichai visited London last week."))

named_entities = ne_chunk(pos_tokens)
print("Named Entities:\n", named_entities)

# To visualize the tree structure
# named_entities.draw() # This might open a new window, might not work in all environments.
```

### 5. Sentiment Analysis
Sentiment analysis determines the emotional tone behind a piece of text. We'll use VADER (Valence Aware Dictionary and sEntiment Reasoner) from `nltk`, which is specifically attuned to sentiments expressed in social media.

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sentiment_analyzer = SentimentIntensityAnalyzer()

# Example sentences
sentences_for_sentiment = [
    "I love this product! It's amazing.",
    "This movie was terrible and boring.",
    "The food was okay, but the service was slow."
]

for sentence in sentences_for_sentiment:
    vs = sentiment_analyzer.polarity_scores(sentence)
    print(f"Sentence: '{sentence}'")
    print(f"Sentiment Scores: {vs}")
    if vs['compound'] >= 0.05:
        print("Overall Sentiment: Positive")
    elif vs['compound'] <= -0.05:
        print("Overall Sentiment: Negative")
    else:
        print("Overall Sentiment: Neutral")
    print("\n")
```
