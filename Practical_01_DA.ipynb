```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Load Dataset
df = pd.read_csv("/content/coffe_sales_data.csv")
#Data Preview
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nShape of dataset:")
print(df.shape)

print("\nDataset info:")
print(df.info())
#Missing Data Analysis
print("\nMissing values in each column:")
print(df.isnull().sum())
#UNIVARIATE ANALYSIS
#(a) Coffee Type Count
df['coffee_name'].value_counts().plot(kind='bar')
plt.title("Coffee Type Count")
plt.xlabel("Coffee Type")
plt.ylabel("Count")
plt.show()
#(b) Transactions by Time of Day
df['Time_of_Day'].value_counts().plot(kind='bar')
plt.title("Transactions by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Count")
plt.show()
#(c) Payment Method Distribution
df['cash_type'].value_counts().plot(kind='bar')
plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
#(d) Revenue Distribution
plt.hist(df['money'], bins=10)
plt.title("Revenue Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
#BIVARIATE ANALYSIS
#(a) Coffee Type vs Time of Day
pd.crosstab(df['coffee_name'], df['Time_of_Day']).plot(kind='bar', stacked=True)
plt.title("Coffee Type vs Time of Day")
plt.xlabel("Coffee Type")
plt.ylabel("Transactions")
plt.show()
#(b) Coffee Type vs Payment Method
pd.crosstab(df['coffee_name'], df['cash_type']).plot(kind='bar', stacked=True)
plt.title("Coffee Type vs Payment Method")
plt.xlabel("Coffee Type")
plt.ylabel("Transactions")
plt.show
#(c) Time of Day vs Average Revenue
df.groupby('Time_of_Day')['money'].mean().plot(kind='bar')
plt.title("Average Revenue by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Average Revenue")
plt.show()
#CORRELATION MATRIX
correlation_matrix = df.corr(numeric_only=True)
print("\nCorrelation Matrix:")
print(correlation_matrix)
#CORRELATION HEATMAP
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
